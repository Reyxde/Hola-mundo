# LENGUAJE python
## Descripción
Python es un lenguaje de programación de alto nivel, interpretado, multiparadigma y de código abierto, creado por Guido van Rossum y lanzado en 1991.  Su filosofía central prioriza la legibilidad y la simplicidad mediante una sintaxis clara que utiliza indentación en lugar de llaves, permitiendo expresar conceptos en menos líneas de código que otros lenguajes como C++ o Java.

# LENGUAJE C++

```python
public class HolaMundo {
    public static void main(String[] args) {
        System.out.println("Hola Mundo");
    }
}
```
